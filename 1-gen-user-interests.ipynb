```python
import pandas as pd

grid_size = 60

records = []
for max_lat in range(90, -90, -grid_size):
    min_lat = max_lat - grid_size
    for min_lon in range(-180, 180, grid_size):
        max_lon = min_lon + grid_size
        records.append([min_lat, max_lat, min_lon, max_lon])

df_grid = pd.DataFrame(records, columns=["min_lat", "max_lat", "min_lon", "max_lon"])
df_grid
```

```python
records = []
for start_year in range(2000, 2020, 10):
    end_year = start_year + 9
    records.append([pd.to_datetime(start_year, format = "%Y"), pd.to_datetime(f"{end_year}-12-31")])
df_year = pd.DataFrame(records, columns=['start_datetime', 'end_datetime'])
df_year
```

```python
df = pd.merge(df_grid, df_year, how='cross')
df
```

```python
def gen_resolution(row):
    min_lat = row["min_lat"]
    start_datetime = row["start_datetime"]
    if start_datetime >= pd.to_datetime("2010", format = "%Y"):
        if min_lat >= 0:
            return "day"
        else:
            return "month"
    else:
        if min_lat >= 0:
            return "month"
        else:
            return "year"

df["resolution"] = df.apply(gen_resolution, axis=1)
df
```

```python
df.to_csv("user_interests.csv", index=False)
```
