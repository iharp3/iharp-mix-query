```python
import pandas as pd

df_ui = pd.read_csv('user_interests.csv')
df_ui
```

```python
def create_meta_record(row):
    new_meta_records = []
    all_resolutions = ["hour", "day", "month", "year"]
    resolution = row.resolution
    for r in range(all_resolutions.index(resolution), len(all_resolutions)):
        res = all_resolutions[r]
        new_meta_records.append([row.min_lat, row.max_lat, row.min_lon, row.max_lon, row.start_datetime, row.end_datetime, res])
    return new_meta_records


meta_records = []
for row in df_ui.itertuples():
    new_records = create_meta_record(row)
    meta_records += new_records

df_meta = pd.DataFrame(
    meta_records, columns=["min_lat", "max_lat", "min_lon", "max_lon", "start_datetime", "end_datetime", "resolution"]
)
# create dummy file name based on index
df_meta["file_name"] = df_meta.index.map(lambda x: f"file_{x}.nc")
df_meta.to_csv("metadata.csv", index=False)
df_meta
```
